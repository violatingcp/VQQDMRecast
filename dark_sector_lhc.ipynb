```python
import uproot
file = uproot.open("relic_vector.root")
#file = uproot.open("/Users/philipharris/Dropbox/Physics/DM/relic/relic_vector.root")

#compute the relic density along a specific ratio of DM and mediator within a tolerance
def relic(iFile,iratio,tolerance=0.02):
    events = iFile['relic']
    vals1=events.arrays(['med','gq1'],'(abs(med-dm/'+str(iratio)+')/med < '+str(tolerance)+') & (gq1 != -5) & (gq1 != -1) & (med < 1250) ',library="np")# & gq1 != -5 & gq1 != -1')
    med=vals1['med'][vals1['med'].argsort()]
    gq=np.exp(vals1['gq1'][vals1['med'].argsort()])
    return med,gq

#make 3d plot
def relic3D(iFile):
    events = iFile['relic']
    vals1=events.arrays(['med','dm','gq1'],'(gq1 != -5) & (gq1 != -1) & (med < 1250) ',library="np")
    med=vals1['med'][vals1['med'].argsort()]
    dm=vals1['dm'][vals1['med'].argsort()]
    gq=np.exp(vals1['gq1'][vals1['med'].argsort()])
    return med,dm,gq

from matplotlib import cm

def plotRelic(iFile,iGMax=1.):
    vmed,vdm,vgq=relic3D(iFile)
    vgq=np.minimum(vgq,iGMax)
    vgq=np.maximum(vgq,1e-3)
    xrange=np.arange(5,305,10)
    yrange=np.arange(5,305,1)
    h1,_,_,_=plt.hist2d(vmed, vdm, bins=[xrange,yrange], weights=np.log10(vgq),cmap=plt.cm.jet)
    #plt.zrange(-3,0)
    plt.xlabel("m$_{dm}$ (GeV)")
    plt.ylabel("$m_med$ (GeV)")
    #plt.zlabel("log_{10} (g$_{q}$)")
    plt.colorbar()
    plt.show()
plotRelic(file)
#gridx=np.arange(0,1000)
#gridy=np.arange(0,1000)
#gridxy=np.meshgrid(gridx,gridy)
#plt.plot(vmed,vdm,vgq,cmap=cm.coolwarm)
#surf = ax.plot_surface(vmed,vdm,vgq, cmap=cm.coolwarm,linewidth=0, antialiased=False)
#print(vmed.shape,vdm.shape,vgq.shape))
#plt.contourf(vmed,vdm,vgq)
#plt.show()
```
